```python




# drop non district data
crops['district'] = crops['district'].str.replace(r'^milenge$', 'milengi', regex=True)
crops['district'] = crops['district'].str.replace(r'^chienge$', 'chiengi', regex=True)
# df['P'] = df['P'].str.replace(r'\D+', '', regex=True).astype('int')

# check district names such as total
def checkdistricts(data1, data2):
    d_list1=[]
    d_list2=[]
    for d in list(data1.district.unique()):
        if d not in list(data2.district.str.lower().unique()):
            d_list1.append(d)
            # print('distiricts not common are:{}'.format(empty_list1))
        else:
            d_list2.append(d)
    return d_list2
# check and replace then drop the non-districts
def cleandistrict(data1,data2):
    d_list=checkdistricts(data1, data2)
    districts_not_common_df=data1[~(data1.district.isin(d_list))]
    d_not_list=list(districts_not_common_df.district.unique())
    data1=data1[~(data1.district.isin(d_not_list))]
    data1 = data1.drop_duplicates()
    return data1
crops=cleandistrict(data1=crops,data2=crops)
crops
```

```python
# drop province
crops=crops.drop('province', axis=1)
```

```python
# clean crops survey data
print(len(crops))
#drop duplicate inputs
crops = crops.drop_duplicates()
print(len(crops))
crops
```

```python
# convert appropriate columns to numeric
crops.planted_ha=pd.to_numeric(crops.planted_ha)
crops.harvested_ha=pd.to_numeric(crops.harvested_ha)
crops.production_mt=pd.to_numeric(crops.production_mt)
crops['yield(mt/ha)']=pd.to_numeric(crops['yield(mt/ha)'])
```

```python

#calculate yield
crops = crops.groupby(['district','year'],as_index=False).sum()
crops['yield'] = crops['production_mt'] / crops['planted_ha']
crops['yield_h'] = crops['production_mt'] / crops['harvested_ha']

```

```python
# #calculate z scores of yield and flag any anomalies
df1 = []
# # crops = list(crops.reset_index())
for d in list(crops.district.unique()):
    df = crops.loc[crops['district']==d].copy()
    df['zyield'] = zscore(df['yield'])
    df1.append(df)
    df2=pd.concat(df1)
```

```python
# add a flag to the dataframe

df2['flag'] = 0
df2['flag'].loc[(((df2['zyield']>1.96).copy())|((df2['zyield']<-1.96).copy()))]= 1
df2['flag'].loc[(((df2['zyield']>2.58).copy())|((df2['zyield']<-2.58).copy()))] = 2
crops_district=df2
crops_district
```
