```python
# !python -m pip install geopandas
import pandas as pd
import numpy as np
import geopandas as gpd
from scipy.stats import zscore
```

```python
#reading rheast and extracting districts
postharvest = pd.read_csv(r'data/postharvest/postharvest_csv')
```

```python
# remove spaces
postharvest.columns=postharvest.columns.str.lower()
postharvest.columns=postharvest.columns.str.strip()
postharvest.columns=postharvest.columns.str.replace(' ', '', regex=True)#remove space in between
```

```python
# remove all white spaces, turn to lower case and
postharvest.province=postharvest.province.str.lower()
postharvest.province = postharvest.province.str.strip()
postharvest.province=postharvest.province.str.replace('-', '', regex=True)#remove hyphen
# postharvest.province=postharvest.province.str.replace(' ', '', regex=True)#removes spaces anywhere

# remove spaces in other columns as well
# postharvest.area_ha = postharvest.area_ha.str.strip()
# postharvest.area_ha=postharvest.area_ha.str.replace(',', '', regex=True)#removes spaces anywhere

# postharvest.production_mt = postharvest.production_mt.str.strip()
# postharvest.production_mt=postharvest.production_mt.str.replace(',', '', regex=True)#removes spaces anywhere

# postharvest.yield = postharvest.yield.str.strip()
# postharvest.yield =postharvest.yield.str.replace(',', '', regex=True)#removes spaces anywhere
```

```python
# drop unnecessary column
postharvest.drop('unnamed:0', axis=1, inplace=True)
postharvest.drop('province.1', axis=1, inplace=True)
postharvest.drop('index.1', axis=1, inplace=True)
postharvest.drop('year.1', axis=1, inplace=True)
postharvest.drop('season.1', axis=1, inplace=True)
```

```python
# clean postharvest survey data
print(len(postharvest))
#drop duplicate inputs
postharvest = postharvest.drop_duplicates()
print(len(postharvest))
postharvest
```

```python
# # convert appropriate columns to numeric
# postharvest.planted_ha=pd.to_numeric(postharvest.planted_ha)
# postharvest.harvested_ha=pd.to_numeric(postharvest.harvested_ha)
# postharvest.production_mt=pd.to_numeric(postharvest.production_mt)
# postharvest['yield(mt/ha)']=pd.to_numeric(postharvest['yield(mt/ha)'])
```

```python

#calculate yield
postharvest = postharvest.groupby(['province','year'],as_index=False).sum()
postharvest['yield'] = postharvest['production_mt'] / postharvest['area_ha']
postharvest

```

```python
# #calculate z scores of yield and flag any anomalies
df1 = []
# # postharvest = list(postharvest.reset_index())
for d in list(postharvest.province.unique()):
    df = postharvest.loc[postharvest['province']==d].copy()
    df['zyield'] = zscore(df['yield'])
    df1.append(df)
    df2=pd.concat(df1)
```

```python
# add a flag to the dataframe
df2['flag'] = 0
df2['flag'].loc[(((df2['zyield']>1.96).copy())|((df2['zyield']<-1.96).copy()))]= 1
df2['flag'].loc[(((df2['zyield']>2.58).copy())|((df2['zyield']<-2.58).copy()))] = 2
postharvest_province=df2
```

```python
# # save to csv
postharvest_province.to_csv('data/postharvest/zambia_postharvest_province_data_clean.csv', index=False)
```
