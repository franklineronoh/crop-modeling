```python
import camelot
import pandas as pd
import os
```

```python
PHS_2011_2012 = 'data/postharvest/PHS_2011_2012.pdf'
PHS_2013_2014 = 'data/postharvest/PHS_2013_2014.pdf'
PHS_2014_2015 = 'data/postharvest/PHS_2014_2015.pdf'
PHS_2012_2013 = 'data/postharvest/PHS_2012_2013.pdf'
# pdfs=os.listdir('data/postharvest')
```

```python

# extract tables
tables_PHS_2011_2012= camelot.read_pdf(PHS_2011_2012, pages='26', lattice=True)
tables_PHS_2012_2013= camelot.read_pdf(PHS_2012_2013, pages='31', lattice=True)
tables_PHS_2014_2015= camelot.read_pdf(PHS_2014_2015, pages='29', lattice=True)
tables_PHS_2014_2015= camelot.read_pdf(PHS_2014_2015, pages='34, 35', lattice=True)
```

```python
# maize crop 2014
# area planted
post_area_2014_2015=tables_PHS_2014_2015[0].df
post_area_2014_2015.rename(columns={0:'province',1:'area_ha',2:'area_percent',3:'basaldressing_mt',
                                    4:'basaldressing_percent',5:'topdressing_mt',6:'topdressing_percent'}, inplace=True)
post_area_2014_2015=post_area_2014_2015[4:]
post_area_2014_2015.reset_index(inplace=True)
post_area_2014_2015['season']='2014_2015['
post_area_2014_2015['year']='2014'


# production
post_production_2014_2015=tables_PHS_2014_2015[1].df
post_production_2014_2015.rename(columns={0:'province',1:'production_mt',2:'production_percent',3:'sold_mt',
                                    4:'sold_percent'}, inplace=True)
post_production_2014_2015=post_production_2014_2015[3:]
post_production_2014_2015.reset_index(inplace=True)
post_production_2014_2015['season']='2014-2015'
post_production_2014_2015['year']='2014'
post_area_2014_2015
```

```python
# maize 2012
post_area_2011_2012=tables_PHS_2011_2012[0].df
post_area_2011_2012.rename(columns={0:'province',1:'area_ha',2:'area_percent',3:'basaldressing_mt',
                                    4:'basaldressing_percent',5:'topdressing_mt',6:'topdressing_percent'}, inplace=True)
post_area_2011_2012=post_area_2011_2012[4:]
post_area_2011_2012.reset_index(inplace=True)
post_area_2011_2012['season']='2011-2012'
post_area_2011_2012['year']='2011'
```

```python
post_area_harvested=pd.concat([post_area_2011_2012,post_area_2012_2013,post_area_2013_2014,post_area_2014_2015], axis=0, ignore_index=True)
post_production=pd.concat([post_production_2011_2012,post_production_2012_2013,post_production_2013_2014,post_production_2014_2015],axis=0, ignore_index=True)
```

```python
# remove all white spaces
post_production.province=post_production.province.str.lower()
post_production.province = post_production.province.str.lstrip()
post_production.province=post_production.province.str.replace('-', '', regex=True)
post_production.province=post_production.province.str.replace(' ', '', regex=True)

post_area_harvested.province=post_area_harvested.province.str.lower()
post_area_harvested.province = post_area_harvested.province.str.lstrip()
post_area_harvested.province=post_area_harvested.province.str.replace('-', '', regex=True)
post_area_harvested.province=post_area_harvested.province.str.replace(' ', '', regex=True)
post_area_harvested.province.unique()
```

```python
# exclude non-provincial data

#Area harvested  
provinces_area=['total', 'totalzambia','zambia']  
area_df=pd.DataFrame(columns=['province', 'area_ha', 'area_percent','basaldressing_mt', 
                              'basaldressing_percent', 'topdressing_mt','topdressing_percent', 'season', 'year'])
for i in provinces_area:
    area_harvested=post_area_harvested.loc[post_area_harvested['province']!=i]
    harvested=pd.concat([area_df,area_harvested], axis=0)
    post_area_harvested=harvested

# production
post_df=pd.DataFrame(columns=['province', 'production_mt', 'production_percent',
       'sold_mt', 'sold_percent', 'season', 'year'])
provinces_production=['total', 'zambia','totalzambia']

for i in provinces_production:
    p=post_production.loc[post_production['province']!=i]
    p_final=pd.concat([post_df,p], axis=0)
    post_production=p_final

# sorted
post_area_sorted=post_area_harvested.sort_values(['province'], ascending=[True], ignore_index=True)
post_production_sorted=post_production.sort_values(['province'], ascending=[True], ignore_index=True)
post_area_harvested=post_area_sorted
post_production=post_production_sorted
```

```python
# combine the area harvest and production data
post=pd.concat([post_area_harvested, post_production], axis=1)
```

```python
# convert area harvested and production to numeric
# remove comma
post.area_ha=post.area_ha.str.lower()
post.area_ha = post.area_ha.str.strip()
post.area_ha=post.area_ha.str.replace(',', '', regex=True)
post.production_mt=post.production_mt.str.lower()
post.production_mt = post.production_mt.str.strip()
post.production_mt=post.production_mt.str.replace(',', '', regex=True)

# convert to numeric
post['production_mt'] = pd.to_numeric(post['production_mt'], errors='coerce')
post['area_ha'] = pd.to_numeric(post['area_ha'], errors='coerce')

# calculate yield
post['yield']=post['production_mt']/post['area_ha']
post
```
